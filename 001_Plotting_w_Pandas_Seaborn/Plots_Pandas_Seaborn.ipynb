## **03. Plot using Seaborn**

```python
path_files= "Datasets/"
```

### **<i>03.1. Line Plot**</i>

**Funcionality:** time series data, price developments, variables that have a behaviour over time. In X axis: variable that evolves in the time, in Y, the observed variable, float!  
**Trick:** Date has to be set as index of dataframe

```python
file_name_line_chart='museum_visitors.csv'
df_line_chart=pd.read_csv(path_files+file_name_line_chart, index_col="Date", parse_dates=True)
df_line_chart.head()
```

```python
#Line plot of all the dataset
sns.lineplot(data=df_line_chart)
```

#Line plot of the dataframe's specific column 
sns.lineplot(data=df_line_chart["Avila Adobe"])

### **<i>03.2. Bar Charts**</i>

```python
file_name_line_chart='ign_scores.csv'
df_bar_chart_heatmap=pd.read_csv(path_files+file_name_line_chart, index_col="Platform")
df_bar_chart_heatmap.head()
```

```python
# VERTICAL Bar chart showing average score for racing games by platform
sns.barplot(x=df_bar_chart_heatmap.index,y=df_bar_chart_heatmap['Racing'])
plt.xticks(rotation=90)
```

```python
# HORIZONTAL Bar chart showing average score for racing games by platform
sns.barplot(x=df_bar_chart_heatmap['Racing'],y=df_bar_chart_heatmap.index)
```

**Note:** to create an horizontal bar chart you only need to change de order of the axis! 

### **<i>03.3. HeatMaps**</i>

```python
# Heatmap showing average game score by platform and genre
sns.heatmap(df_bar_chart_heatmap, annot=True)
```
