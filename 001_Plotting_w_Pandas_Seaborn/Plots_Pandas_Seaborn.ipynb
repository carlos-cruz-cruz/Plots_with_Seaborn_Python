### **<i>03.4. Scatter Plots**</i>

**What is it?**: A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.  

**Uses**: 
- Observe and show relationships between two numeric variables.  
- Identification of correlational relationships.  
- It is useful for identifying other patterns in the data, such as dividing data points into groups based on the proximity of groups of points.  
- Can also show if there are any unexpected gaps in the data and if there are any outlier points.  
- Useful if we want to segment the data into different parts.  r.

```python
# Load the data
file_name_scatter_plot='insurance.csv'
insurance_data=pd.read_csv(path_files+file_name_scatter_plot)
insurance_data.head()
```

**Simple Scatter Plot**: just plots the points of 2 variables.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

**Simple Scatter Plot with Regression line**: It plots the points of 2 variables and add a regression line.

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

**Scatter Plot for 3 variables - Only points**: display the relationships between three variables, by color-coding the points.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

**Note:** in this case, the dots change color according to the value of the variable marked in "hue", and automatically, the legends of this variable will appear on the plot.

**Scatter Plot for 3 variables - With Regression Line**: display the relationships between three variables, by color-coding the points, and add a regression line for each value of the variable in "hue".

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

**Note:** you could note that the <i>sns.lmplot</i> command works slightly differently than the other scatterplots commands, due that instead of setting <i>x=insurance_data['bmi']</i> to select the 'bmi' column in insurance_data, we set <i>x="bmi"</i> to specify the name of the column only.
Similarly, <i>y="charges"</i> and <i>hue="smoker"</i> also contain the names of columns. We specify the dataset with data=insurance_data.

**Categorical scatter plot**: scatter plots to highlight the relationship between two variables, one continuous and one categorical.

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```

### **<i>03.5. Histograms**</i>

**What is it?**: A histogram is a chart that plots the distribution of a numeric variable‚Äôs values as a series of bars. Each bar typically covers a range of numeric values called a bin or class; a bar‚Äôs height indicates the frequency of data points with a value within the corresponding bin.  

**Uses**: 
- Only for variables that takes <u>continuous numeric values.</u>
- Good for showing general distributional features of dataset variables: Identify peaks, check skewness and kurtosis.
- Identifying outliers

Information about the number of bins and their boundaries for tallying up the data points is not inherent to the data itself. Instead, setting up the bins is a separate decision that we have to make when constructing a histogram. The way that we specify the bins will have a major effect on how the histogram can be interpreted, as will be seen below.


```python
# Load the data
file_name_hits_DP='iris.csv'
iris_data=pd.read_csv(path_files+file_name_hits_DP, index_col="Id")
iris_data.head()
```

**Regular Histogram**: automatic numbers of bins. 

```python
# Histogram - Variable on horizontal axis
sns.histplot(data=iris_data, x='Petal Length (cm)') # For get a vertical histogram, use y instead of x.
```

**Note:** the number of bins is automatically calculated by Seaborn, using numpy.histogram_bin_edges. For small datasets the Sturges value will usually be chosen, while larger datasets will usually default to Freedman Diaconis Estimator (FD):


**‚ÄòFd‚Äô (Freedman Diaconis Estimator):**  
The binwidth is proportional to the interquartile range (IQR) and inversely proportional to cube root of a.size. Can be too conservative for small datasets, but is quite good for large datasets. The IQR is very robust to outliers.  

$ h = 2 \cdot \frac{\text{IQR}(x)}{\sqrt[3]{n}}$  

**‚ÄòSturges‚Äô Estimator:**  
The number of bins is the base 2 log of a.size. This estimator assumes normality of data and is too conservative for larger, non-normal datasets. This is the default method in R‚Äôs hist method.  

$ k = \lceil \log_2(n) + 1 \rceil $

However, if neccesary, you can change the method to use, could be:

**‚ÄòScott‚Äô:**  
$ k = \lceil \frac{3.5 \cdot \text{std}(x)}{\sqrt[3]{n}} \rceil $

**‚ÄòRice‚Äô:** 

$ k = \lceil 2 \cdot n^{1/3} \rceil $

You can see others in: https://numpy.org/doc/stable/reference/generated/numpy.histogram_bin_edges.html#numpy.histogram_bin_edges

**Histogram with a KDE**: you can add a kernel density estimate to smooth the histogram, providing complementary information about the shape of the distribution.

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', kde=True)
```

**Histogram with a Hue**: You can otherwise draw multiple histograms from a long-form dataset with hue mapping. The default approach to plotting multiple distributions is to ‚Äúlayer‚Äù them, but you can also ‚Äústack‚Äù, using multiple="stack", default ‚Äúlayer‚Äù.

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', hue="Species", multiple="stack")
```

**Histogram for two continuous variables**:

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', y="Sepal Width (cm)",cbar=True,)
```

### **<i>03.6. Density plots - KDE:**</i>

**What is it?**: A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions. A histogram aims to approximate the underlying probability density function that generated the data by binning and counting observations. Kernel density estimation (KDE) presents a different solution to the same problem. Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate.

**Uses**: 
- Visualizing the distribution of observations in a dataset
- KDE can produce a plot that is less cluttered and more interpretable, especially when drawing multiple distributions
- Has the potential to introduce distortions if the underlying distribution is bounded or not smooth

For this plots, we continue to use the same data as in numeral 03.5.

**Regular KDE plot for one varaible:**

```python
sns.kdeplot(data=iris_data['Petal Length (cm)'], fill=True) # With fill a shadow is created under the plot.
```

**Regular KDE plot for several variables:**

```python
sns.kdeplot(data=iris_data)
```

**KDE plot for one variable with Hue:**

```python
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue="Species")
```

**Cumulative Distribution Function one variable with Hue:**

```python
sns.kdeplot(
    data=iris_data, x='Petal Length (cm)', hue="Species",
    cumulative=True, common_norm=False, common_grid=True,
)
```

**2D KDE plots:**

```python
# 2D KDE plot
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde",fill=True,cmap='Blues') 
#You can remove fill, and you'll get a line surface. Cmap is the color palette.
```

**Note:** In a 2D KDE plot, density is typically represented as a smooth surface that rises above the xy-plane. The higher the density value at a given point on the plane, the greater the concentration of data in that region, and additionally, you can see a regular one variable KDE up and right side, exposing the behavior of each variable.
